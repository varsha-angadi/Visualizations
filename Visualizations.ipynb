```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Load the dataset
AFL = pd.read_csv(r'C:\Users\varsh\OneDrive\Documents\CSV files\AFL.csv')
BAC = pd.read_csv(r'C:\Users\varsh\OneDrive\Documents\CSV files\BAC.csv')
CB = pd.read_csv(r'C:\Users\varsh\OneDrive\Documents\CSV files\CB.csv')
GS = pd.read_csv(r'C:\Users\varsh\OneDrive\Documents\CSV files\GS.csv')
WFC = pd.read_csv(r'C:\Users\varsh\OneDrive\Documents\CSV files\WFC.csv')
```

```python
# Summary of the datasets
AFL.info()
BAC.info()
CB.info()
GS.info()
WFC.info()
```

```python
AFL['Date']=pd.to_datetime(AFL['Date'])
BAC['Date']=pd.to_datetime(BAC['Date'])
CB['Date']=pd.to_datetime(CB['Date'])
GS['Date']=pd.to_datetime(GS['Date'])
WFC['Date']=pd.to_datetime(WFC['Date'])
```

```python
AFL.info()
```

```python
AFL.describe()
```

```python
tickers = ['AFL', 'BAC', 'CB', 'GS', 'WFC']
```

```python
bank_stocks = pd.concat([AFL, BAC, CB, GS, WFC], axis=1, keys=tickers)
# bank_stocks
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
bank_stocks.head()
```

```python
bank_stocks.info()
```

```python
bank_stocks.xs('Close', axis=1, level='Stock Info').max()
```

```python
WFC['Close'].idxmax()
```

```python
# Plotting with Matplotlib

# Line Plot
plt.figure(figsize=(12, 6))
plt.plot(BAC['Date'], BAC['Close'])
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(12, 6))
plt.bar(AFL['Date'], AFL['Volume'])
plt.title('Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram
plt.figure(figsize=(12, 6))
plt.hist(CB['Close'], bins=50)
plt.title('Distribution of Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(12, 6))
plt.scatter(GS['Open'], GS['Close'])
plt.title('Opening vs Closing Prices')
plt.xlabel('Opening Price')
plt.ylabel('Closing Price')
plt.show()
```

```python
# Plotting with Seaborn

# Line Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Close', data=WFC)
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting with Plotly

# Line Plot
fig = px.line(GS, x='Date', y='Close', title='Stock Prices Over Time')
fig.show()
```

```python
# Bar Plot
fig = px.bar(BAC, x='Date', y='Volume', title='Trading Volume Over Time')
fig.show()
```

```python
# Histogram
fig = px.histogram(WFC, x='Close', nbins=50, title='Distribution of Closing Prices')
fig.show()
```

```python
# Scatter Plot
fig = px.scatter(CB, x='Open', y='Close', title='Opening vs Closing Prices')
fig.show()
```
